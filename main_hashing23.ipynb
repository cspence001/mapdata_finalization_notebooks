```python
ppp_df_23_hash["NAICSCode"] = ppp_df_23_hash["NAICSCode"].astype(str)
ppp_df_23_hash["NAICSCode"] = ppp_df_23_hash.apply(lambda x: x["NAICSCode"] + '00000' if x["NAICSCode"]=='0' else x["NAICSCode"], axis=1)
ppp_df_23_hash.head()
```

```python
ppp_df_23_hash.loc[ppp_df_23_hash["NAICSCode"].str.len() < 6]
```

```python
ppp_df_23_hash["INDcode"] = 'IND' + ppp_df_23_hash["NAICSCode"].astype(str).str[:2]
ppp_df_23_hash["SScode"] = 'SS' + ppp_df_23_hash["NAICSCode"].astype(str).str[2:4]
ppp_df_23_hash.head()
```

Loan Conditions

```python
loan_conditions = [
(ppp_df_23_hash.LoanRange == "a $150,000-350,000" ),
(ppp_df_23_hash.LoanRange == "b $350,000-1 million" ),
(ppp_df_23_hash.LoanRange == "c $1-2 million" ),
(ppp_df_23_hash.LoanRange == "d $2-5 million" ),
(ppp_df_23_hash.LoanRange == "e $5-10 million" ),
]
loan_values = ["LR1","LR2","LR3","LR4","LR5"]
ppp_df_23_hash['LRcode'] = np.select(loan_conditions, loan_values)
ppp_df_23_hash.head()
```

```python
jobrange_conditions = [
    (ppp_df_23_hash.JobRange == 'Fewer than 5'),
    (ppp_df_23_hash.JobRange == '5 to 9'),
    (ppp_df_23_hash.JobRange == '10 to 19'),
    (ppp_df_23_hash.JobRange == '20 to 49'),
    (ppp_df_23_hash.JobRange == '50 to 99'),
    (ppp_df_23_hash.JobRange == '100 to 249'),
    (ppp_df_23_hash.JobRange == '250 to 499'),
    (ppp_df_23_hash.JobRange == '500 or more'),
]

jobrange_values = ['JR1', 'JR2', 'JR3', 'JR4', 'JR5',
       'JR6', 'JR7', 'JR8']

ppp_df_23_hash['JRcode'] = np.select(jobrange_conditions , jobrange_values)
ppp_df_23_hash.head()
```

```python
businessage_conditions = [
    (ppp_df_23_hash.BusinessAge == 'Existing or more than 2 years old'),
    (ppp_df_23_hash.BusinessAge == 'New Business or 2 years or less'),
    (ppp_df_23_hash.BusinessAge == 'Not Available'),
    (ppp_df_23_hash.BusinessAge == 'Change of Ownership'),
    (ppp_df_23_hash.BusinessAge == 'Startup, Loan Funds will Open Business'),
]
businessage_values = ["BA1", "BA2", "BA3", "BA4", "BA5"]
ppp_df_23_hash['BAcode'] = np.select(businessage_conditions, businessage_values)
ppp_df_23_hash.head()
```

```python
ppp_df_23_hash["PMcode"] = ppp_df_23_hash["ProcessingMethod"] # PPP/PPS
ppp_df_23_hash["NPcode"] = 'NP' + ppp_df_23_hash["NonProfit"] # Y/N
ppp_df_23_hash["LMIcode"] = 'LMI' +ppp_df_23_hash["LMIIndicator"] #Y/N
ppp_df_23_hash["HZcode"] = 'HZ' + ppp_df_23_hash["HubzoneIndicator"] #Y/N
ppp_df_23_hash["RUcode"] = ppp_df_23_hash["RuralUrbanIndicator"] #U/R
ppp_df_23_hash.head()
```
