<h4>STATE PROCESS CHECKS

```python
path21_hpgf =pd.read_csv("../../data/pipe_hpgf.csv", index_col=0,dtype={'ForgivenessAmount': str, 'UndisbursedAmount': str})
path23_hpgf =pd.read_csv("../../data/pipe_hpgf23.csv", index_col=0,dtype={'ForgivenessAmount': str, 'UndisbursedAmount': str})
```

```python
set(path21_hpgf.columns).symmetric_difference(path23_hpgf.columns)
```

<b>21 GEO COUNTS BY STATE

```python
cols = ["FullAdd","Lat","Long","FIPS_z"]
null21= path21_hpgf[cols].isnull()
null21_vc=null21.value_counts()
null21_vc
```

```python
path21_hpgf_grp= path21_hpgf.groupby('BorrowerState')[['FullAdd','Lat','Long','FIPS_z']].apply(lambda x: x.isna().sum())
state_counts21 = path21_hpgf["BorrowerState"].value_counts()
state_countsdf21 = pd.DataFrame({'BorrowerState': state_counts21.index, 'count':state_counts21.values})
merge_21= state_countsdf21.merge(path21_hpgf_grp, on='BorrowerState', how='left')
merge_21= merge_21.sort_values(["BorrowerState"], ascending=True)
merge_21
```

<b>21 NULL COLUMN COUNTS BY STATE

```python
pd.set_option('display.max_columns', None)
path21_hpgf_nullcnts= path21_hpgf[path21_hpgf.columns.difference(['BorrowerState'])].isnull().groupby(path21_hpgf.BorrowerState).sum().astype(int)
path21_hpgf_drop0= path21_hpgf_nullcnts.loc[:, path21_hpgf_nullcnts.any()] #drops columns if all values are False
path21_hpgf_drop0
```

<b>23 GEO COUNTS BY STATE

```python
cols = ["FullAdd","Lat","Long","FIPS_z"]
null23= path23_hpgf[cols].isnull()
null23_vc=null23.value_counts()
null23_vc
```

```python
path23_hpgf_grp= path23_hpgf.groupby('BorrowerState')[['FullAdd','Lat','Long','FIPS_z']].apply(lambda x: x.isna().sum())
state_counts23 = path23_hpgf["BorrowerState"].value_counts()
state_countsdf23 = pd.DataFrame({'BorrowerState': state_counts23.index, 'count':state_counts23.values})
merge_23= state_countsdf23.merge(path23_hpgf_grp, on="BorrowerState", how='left')
merge_23
```

<b>23 NULL COLUMN COUNTS BY STATE

```python
pd.set_option('display.max_columns', None)
path23_hpgf_nullcnts= path23_hpgf[path23_hpgf.columns.difference(['BorrowerState'])].isnull().groupby(path23_hpgf.BorrowerState).sum().astype(int)
path23_hpgf_drop0= path23_hpgf_nullcnts.loc[:, path23_hpgf_nullcnts.any()] #drops columns if all values are False
path23_hpgf_drop0
```

<b>2021, 2023 Nulls, Not Available

```python
path21_hpgf_nullcol= path21_hpgf[path21_hpgf.columns[path21_hpgf.isnull().any()]].isnull().sum()
path21col_df = pd.DataFrame({'column_nu': path21_hpgf_nullcol.index, 'count_21':path21_hpgf_nullcol.values.astype(str)})
path21col_df = path21col_df.sort_values(["column_nu"], ascending=True)

na = path21_hpgf[path21_hpgf == 'Not Available'].count()
na0 =na[na!=0]
na0_df= pd.DataFrame({'column_na': na0.index, 'count_21':na0.values.astype(str)})
na0_df=na0_df.sort_values(["column_na"], ascending=True)

frames21 = [path21col_df,na0_df]
result21 = pd.concat(frames21)
result21
```

```python
path23_hpgf_nullcol = path23_hpgf[path23_hpgf.columns[path23_hpgf.isnull().any()]].isnull().sum()
path23col_df = pd.DataFrame({'column_nu': path23_hpgf_nullcol.index, 'count_23':path23_hpgf_nullcol.values})
path23col_df = path23col_df.sort_values(["column_nu"], ascending=True)

na23 = path23_hpgf[path23_hpgf == 'Not Available'].count()
na023 =na23[na23!=0]
na023_df = pd.DataFrame({'column_na': na023.index, 'count_23':na023.values})
na023_df=na023_df.sort_values(["column_na"], ascending=True)

frames23 = [path23col_df,na023_df]
result23 = pd.concat(frames23)
result23 
```

```python
result_merge=result21.merge(result23, on=["column_na","column_nu"], how="outer")
result_merge
```
