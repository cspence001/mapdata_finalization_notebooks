Determining Necessary Columns for Final DataFrame

```python
#keep track of columns removed
removed = ['NAICSCode', 'Industry', 'IndustrySubsector', 'LoanRange', 'JobRange', 'RuralUrbanIndicator','HubzoneIndicator','LMIIndicator']
#Industry and Industry Subsector taken from HASH file (NAICS)

cols =['LoanNumber','DateApprovedTS','SBAOfficeCode','SBAOfficeLabel','ProcessingMethod','BorrowerName',
       'LoanStatusDateTS','LoanStatus','Term','CurrentApprovalAmount','UndisbursedAmount', 
       'ServicingLenderLocationID','ServicingLenderName','ServicingLenderState','ServicingLenderFullAddress',
       'EstimatedLenderPercent', 'EstimatedLenderProfit', 
       'BusinessType','BusinessAgeDescription','CD','JobsReported','NAICSCode','LongName',
       'OriginatingLenderLocationID','OriginatingLender', 'OriginatingLenderCity', 'OriginatingLenderState',
       'NonProfit', 'ForgivenessAmount', 'ForgivenessDateTS'] #30

ppp_pipe23 = ppp_pipe23[cols]
ppp_pipe23.head()
```

<b>HASH file addition

```python
ppp_hash = pd.read_csv("../data/ppp_hashSort23.csv") 
ppp_hash = ppp_hash.drop(["Unnamed: 0"], axis=1)
ppp_hash
```

<b>GEO, FIPS file addition  

```python
ppp_geo_fips=pd.read_csv("../data/state_data/geo/ppp_geo_fips23.csv", index_col=0)
ppp_geo_fips
```

<b>Merged DataFrame

HASH > MAIN 23

```python
#merge 
pd.set_option('display.max_columns', None)
ppp_hash_pipe23 = ppp_pipe23.merge(ppp_hash, on="LoanNumber", how="left")
ppp_hash_pipe23
```

HASH PIPE > RESULT (FIPS, GEO)

```python
# Combine (coords, fips) only if "LoanNumber", "BorrowerState", "full_add" match 
ppp_hpgf = ppp_hash_pipe23.merge(ppp_geo_fips, on=["LoanNumber"], how="left")
ppp_hpgf
```

```python
#checks
pd.set_option('display.max_rows', 70)
print(ppp_hpgf.isnull().sum())
```
