<h3>Hashing Code Values

```python
pd.set_option('display.max_columns', None)
ppp_df_23_hash = ppp_df_23
ppp_df_23_hash = ppp_df_23_hash.drop(["Unnamed: 0"], axis=1)
ppp_df_23_hash.head()
```

Industry Conditions

```python
ppp_df_23_hash["NA"]=ppp_df_23_hash["NA"].apply(str)
ppp_df_23_hash["4DigitNAICS"]=ppp_df_23_hash["4DigitNAICS"].apply(str)
```

```python
ppp_df_23_hash["INDcode"] = 'IND' + ppp_df_23_hash["NA"]
ppp_df_23_hash["SScode"] = 'SS' + ppp_df_23_hash["4DigitNAICS"].str[-2:]
ppp_df_23_hash.head()
```

Loan Conditions

```python
loan_conditions = [
(ppp_df_23_hash.Loan_Range == "a $150,000-350,000" ),
(ppp_df_23_hash.Loan_Range == "b $350,000-1 million" ),
(ppp_df_23_hash.Loan_Range == "c $1-2 million" ),
(ppp_df_23_hash.Loan_Range == "d $2-5 million" ),
(ppp_df_23_hash.Loan_Range == "e $5-10 million" ),
]
loan_values = ["LR1","LR2","LR3","LR4","LR5"]
ppp_df_23_hash['LRcode'] = np.select(loan_conditions, loan_values)
ppp_df_23_hash.head()
```

Job Range Conditions

```python
jobrange_conditions = [
    (ppp_df_23_hash.job_Range == 'Fewer than 5'),
    (ppp_df_23_hash.job_Range == '5 to 9'),
    (ppp_df_23_hash.job_Range == '10 to 19'),
    (ppp_df_23_hash.job_Range == '20 to 49'),
    (ppp_df_23_hash.job_Range == '50 to 99'),
    (ppp_df_23_hash.job_Range == '100 to 249'),
    (ppp_df_23_hash.job_Range == '250 to 499'),
    (ppp_df_23_hash.job_Range == '500 or more'),
]

jobrange_values = ['JR1', 'JR2', 'JR3', 'JR4', 'JR5',
       'JR6', 'JR7', 'JR8']

ppp_df_23_hash['JRcode'] = np.select(jobrange_conditions , jobrange_values)
ppp_df_23_hash.head()
```

```python
ppp_df_23_hash['JRcode'].value_counts()
```

Business Type Conditions

```python
businesstype_conditions = [
    (ppp_df_23_hash.BusinessType == "Corporation"),
    (ppp_df_23_hash.BusinessType == "Limited Liability Company(LLC)"),
    (ppp_df_23_hash.BusinessType == "Subchapter S Corporation"),
    (ppp_df_23_hash.BusinessType == "Non-Profit Organization"),
    (ppp_df_23_hash.BusinessType == "Partnership"),
    (ppp_df_23_hash.BusinessType == "Limited Liability Partnership"),
    (ppp_df_23_hash.BusinessType == "Sole Proprietorship"),
    (ppp_df_23_hash.BusinessType == "Professional Association"),
    (ppp_df_23_hash.BusinessType == "Cooperative"),
    (ppp_df_23_hash.BusinessType == "501(c)3 – Non Profit"),
    (ppp_df_23_hash.BusinessType == "Non-Profit Childcare Center"),
    (ppp_df_23_hash.BusinessType == "501(c)6 – Non Profit Membership"),
    (ppp_df_23_hash.BusinessType == "Not Available"),
    (ppp_df_23_hash.BusinessType == "Employee Stock Ownership Plan(ESOP)"),
    (ppp_df_23_hash.BusinessType == "Self-Employed Individuals"),
    (ppp_df_23_hash.BusinessType == "Trust"),
    (ppp_df_23_hash.BusinessType == "Housing Co-op"),
    (ppp_df_23_hash.BusinessType == "Joint Venture"),
    (ppp_df_23_hash.BusinessType == "Independent Contractors"),
    (ppp_df_23_hash.BusinessType == "Single Member LLC"),
    (ppp_df_23_hash.BusinessType == "Tribal Concerns"),
    (ppp_df_23_hash.BusinessType == "Tenant in Common"),
    (ppp_df_23_hash.BusinessType == "501(c) – Non Profit except 3,4,6,"),
    (ppp_df_23_hash.BusinessType == "Rollover as Business Start-Ups (ROB"),
    (ppp_df_23_hash.BusinessType == "501(c)19 – Non Profit Veterans"),
    (ppp_df_23_hash.BusinessType == "Qualified Joint-Venture (spouses)"),
    
    
]

businesstype_values = ["BT01", "BT02", "BT03", "BT04", "BT05", 
                       "BT06", "BT07", "BT08", "BT09", "BT10", 
                       "BT11", "BT12", "BT13", "BT14", "BT15", 
                       "BT16", "BT17", "BT18", "BT19", "BT20", 
                       "BT21", "BT22", "BT23", "BT24", "BT25",
                       "BT26"]
ppp_df_23_hash['BTcode'] = np.select(businesstype_conditions, businesstype_values)
ppp_df_23_hash.head()
```

```python
ppp_df_23_hash['BTcode'].value_counts()
```

Business Age Conditions

```python
businessage_conditions = [
    (ppp_df_23_hash.BusinessAgeDescription == 'Existing or more than 2 years old'),
    (ppp_df_23_hash.BusinessAgeDescription == 'New Business or 2 years or less'),
    (ppp_df_23_hash.BusinessAgeDescription == 'Unanswered'),
    (ppp_df_23_hash.BusinessAgeDescription == 'Change of Ownership'),
    (ppp_df_23_hash.BusinessAgeDescription == 'Startup, Loan Funds will Open Business'),
]
businessage_values = ["BA1", "BA2", "BA3", "BA4", "BA5"]
ppp_df_23_hash['BAcode'] = np.select(businessage_conditions, businessage_values)
ppp_df_23_hash.head()
```
