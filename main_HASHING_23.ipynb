<h4>Hashing Code Values

```python
pd.set_option('display.max_columns', None)
ppp_df_hash = ppp_df
ppp_df_hash = ppp_df_hash.drop(["Unnamed: 0", "Unnamed: 0.1"], axis=1)
ppp_df_hash.head()
```

Industry Conditions

```python
ppp_df_hash["NA"]=ppp_df_hash["NA"].apply(str)
ppp_df_hash["4DigitNAICS"]=ppp_df_hash["4DigitNAICS"].apply(str)
```

```python
ppp_df_hash["INDcode"] = 'IND' + ppp_df_hash["NA"]
ppp_df_hash["SScode"] = 'SS' + ppp_df_hash["4DigitNAICS"].str[-2:]
ppp_df_hash.head()
```

Loan Conditions

```python
loan_conditions = [
(ppp_df_hash.Loan_Range == "a $150,000-350,000" ),
(ppp_df_hash.Loan_Range == "b $350,000-1 million" ),
(ppp_df_hash.Loan_Range == "c $1-2 million" ),
(ppp_df_hash.Loan_Range == "d $2-5 million" ),
(ppp_df_hash.Loan_Range == "e $5-10 million" ),
]
loan_values = ["LR1","LR2","LR3","LR4","LR5"]
ppp_df_hash['LRcode'] = np.select(loan_conditions, loan_values)
ppp_df_hash.head()
```

Job Range Conditions

```python
jobrange_conditions = [
    (ppp_df_hash.job_Range == 'Fewer than 5'),
    (ppp_df_hash.job_Range == '5 to 9'),
    (ppp_df_hash.job_Range == '10 to 19'),
    (ppp_df_hash.job_Range == '20 to 49'),
    (ppp_df_hash.job_Range == '50 to 99'),
    (ppp_df_hash.job_Range == '100 to 249'),
    (ppp_df_hash.job_Range == '250 to 499'),
    (ppp_df_hash.job_Range == '500 or more'),
]

jobrange_values = ['JR1', 'JR2', 'JR3', 'JR4', 'JR5',
       'JR6', 'JR7', 'JR8']

ppp_df_hash['JRcode'] = np.select(jobrange_conditions , jobrange_values)
ppp_df_hash.head()
```
