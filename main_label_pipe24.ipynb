<h5> 2023 to 2024 Raw Data Null Comp

```python
#2023 Raw Data for Comp
csv_path_23 = "../data/raw/>150K/public_150k_plus_230630.csv"
ppp_df_23 = pd.read_csv(csv_path_23, encoding='iso-8859-1')
```

```python
pd.set_option('display.max_rows', None)
# Calculate null counts
null_counts_23 = ppp_df_23.isnull().sum().rename("Nulls_2023")
null_counts_24 = ppp_df_24.isnull().sum().rename("Nulls_2024")

# Combine into a single DataFrame
null_comparison_df = pd.concat([null_counts_23, null_counts_24], axis=1)

# Add difference column (2024 - 2023)
null_comparison_df["Difference"] = null_comparison_df["Nulls_2024"] - null_comparison_df["Nulls_2023"]

# Reset index to show column names as a column
null_comparison_df = null_comparison_df.reset_index().rename(columns={"index": "Column"})

# Display the DataFrame
print(null_comparison_df)

# Optional: Save to CSV
null_comparison_df.to_csv("../data/null/null_counts_comparison_2023_2024.csv", index=False)
```

<b>2024 Raw Data Null Report

```python
rows_with_nulls = ppp_df_24[ppp_df_24.isnull().any(axis=1)]
cols_with_nulls = ppp_df_24.columns[ppp_df_24.isnull().any()]
result = rows_with_nulls[cols_with_nulls]

# Apply highlighting to ONLY the null-containing subset
styled_result = result.style.highlight_null(color='yellow')

# display(styled_result)
with open("../data/null/highlighted_nulls_2024.html", "w") as f:
    f.write(styled_result.to_html())
```

<b>Null Address Update

```python
# Define columns to compare
cols = ['LoanNumber', 'BorrowerName', 'BorrowerAddress', 'BorrowerCity', 'BorrowerState', 'BorrowerZip']
#Get null addresses from 2024 data
null_address_df = ppp_df_24.loc[ppp_df_24[cols[2:]].isnull().any(axis=1), cols]
print(f"Found {len(null_address_df)} null addresses (address, city, state, and/or zip) loan records")
display(null_address_df)

# 2024 Data Null Address Report
# Apply highlighting to ONLY the null-containing subset
null_address_df24_styled = null_address_df.style.highlight_null(color='yellow')

# display(styled_result)
with open("../data/null/highlighted_nulls_addresses_2024.html", "w") as f:
    f.write(null_address_df24_styled.to_html())
```

```python
#Read in address updated 2023 data
ppp_df_23_address = pd.read_csv("../data/raw/>150K/public_150k_plus_230630_add.csv", index_col=0)
ppp_df_23_address.head()
```

<b> SxS 2023 x 2024 Address Null Comp

```python
# SxS Diff
null_loan_numbers = null_address_df['LoanNumber']
#Filter the previous year's data for matching loans
matching_addresses = ppp_df_23_address[ppp_df_23_address['LoanNumber'].isin(null_loan_numbers)][cols]  # Using the same columns as before
print(f"Found {len(matching_addresses)} matching loan records")

#Merge for side-by-side comparison
comparison_df = pd.merge(
    null_address_df,
    matching_addresses,
    on='LoanNumber',
    how='left',
    suffixes=('_2024', '_2023'))
    
#Highlight differences
def highlight_diff(row):
    styles = [''] * len(row)
    for i, col in enumerate(cols[1:]):  # Skip LoanNumber
        if row[f'{col}_2024'] != row[f'{col}_2023']:
            styles[i+1] = 'background-color: yellow'
    return styles

styled_comparison = comparison_df.style.apply(highlight_diff, axis=1)
display(styled_comparison)
```

<b>Naming Consistency: Borrower Name

```python
# Clean &AMP; in both DataFrames before processing
# Create a report of affected names before cleaning
amp_report = ppp_df_24[ppp_df_24['BorrowerName'].str.contains('&AMP;', na=False)][['LoanNumber', 'BorrowerName']].copy()
amp_report['CleanedName'] = amp_report['BorrowerName'].str.replace('&AMP;', '&')

# Clean both DataFrames
ppp_df_24['BorrowerName'] = ppp_df_24['BorrowerName'].str.replace('&AMP;', '&')
ppp_df_23_address['BorrowerName'] = ppp_df_23_address['BorrowerName'].str.replace('&AMP;', '&')

ppp_df_24.loc[ppp_df_24["LoanNumber"] == 9712177706, 'BorrowerName'] = "COMPLETE WELDING & FABRICATION INC"

# Save AMP cleaning report
amp_report.to_html('../data/null/reports/df20240930_amp_cleaning_report.html', index=False)

cleanamp_check = ppp_df_24[ppp_df_24['BorrowerName'].astype(str).str.contains("&AMP;")]
cleanamp_check.head()
```

```python
# Update Code

address_cols = ['BorrowerAddress', 'BorrowerCity', 'BorrowerState', 'BorrowerZip']
report_datetime = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

# Create a copy of the original data for reporting
original_values = ppp_df_24[ppp_df_24[address_cols].isnull().any(axis=1)].copy()

# Get indices of rows with null addresses
null_indices = ppp_df_24[ppp_df_24[address_cols].isnull().any(axis=1)].index

# Update values directly in ppp_df_24
updated_records = []
for idx in null_indices:
    row = ppp_df_24.loc[idx]
    match = ppp_df_23_address[
        (ppp_df_23_address['LoanNumber'] == row['LoanNumber']) &
        (ppp_df_23_address['BorrowerName'] == row['BorrowerName'])
    ]
    
    changes = {}
    if not match.empty:
        for col in address_cols:
            if pd.isnull(row[col]) and not pd.isnull(match[col].values[0]):
                # Store original value for reporting
                original_value = row[col]
                # Update the main DataFrame
                ppp_df_24.at[idx, col] = match[col].values[0]
                changes[col] = {
                    'from': 'NULL' if pd.isnull(original_value) else original_value,
                    'to': match[col].values[0]
                }
    
    if changes:
        updated_records.append({
            'LoanNumber': row['LoanNumber'],
            'BorrowerName': row['BorrowerName'],
            'changes': changes
        })

# Generate HTML report (same as before)
html_content = f"""
<html>
<head>
    <title>Address Field Updates Report</title>
    <style>
        body {{ font-family: Arial, sans-serif; margin: 20px; }}
        h1 {{ color: #2c3e50; }}
        .report-info {{ margin-bottom: 20px; }}
        table {{ border-collapse: collapse; width: 100%; margin-bottom: 20px; }}
        th, td {{ border: 1px solid #ddd; padding: 8px; text-align: left; }}
        th {{ background-color: #f2f2f2; }}
        .null {{ color: #999; font-style: italic; }}
        .changed {{ background-color: #fffacd; }}
        .summary {{ background-color: #f8f9fa; padding: 15px; border-radius: 5px; }}
    </style>
</head>
<body>
    <h1>Address Field Updates Report</h1>
    <div class="report-info">
        <p><strong>Report generated:</strong> {report_datetime}</p>
        <p><strong>Total records with nulls:</strong> {len(null_indices)}</p>
        <p><strong>Records updated:</strong> {len(updated_records)}</p>
        <p><strong>Source file:</strong> public_150k_plus_230630_add.csv</p>
        <p><strong>Target file:</strong> public_150k_plus_240930.csv</p>
    </div>
    <table>
        <tr>
            <th>LoanNumber</th>
            <th>BorrowerName</th>
            <th>Field</th>
            <th>Original Value</th>
            <th>Updated Value</th>
        </tr>
"""

for record in updated_records:
    for field, change in record['changes'].items():
        html_content += f"""
        <tr>
            <td>{record['LoanNumber']}</td>
            <td>{record['BorrowerName']}</td>
            <td>{field}</td>
            <td class="null">{change['from']}</td>
            <td class="changed">{change['to']}</td>
        </tr>
        """

html_content += """
    </table>
</body>
</html>
"""

# Save HTML report
report_filename = f"df09302024_address_updates_report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.html"
report_path = f"../data/null/reports/{report_filename}"
with open(report_path, "w") as f:
    f.write(html_content)

print(f"Updated {len(updated_records)} address records in main DataFrame")
print(f"HTML report saved to: {report_path}")

# Verify updates in main DataFrame
print("\nSample of updated records in main DataFrame:")
display(ppp_df_24.loc[null_indices, ['LoanNumber', 'BorrowerName'] + address_cols].head())
```
